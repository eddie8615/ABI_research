```python
import pandas as pd
import os
```

```python
data_path = os.path.relpath('./Data')
```

```python
df_train = pd.read_csv(os.path.join(data_path, 'train.csv'))
df_validation = pd.read_csv(os.path.join(data_path, 'validation.csv'))
df_test = pd.read_csv(os.path.join(data_path, 'test.csv'))

df = pd.concat([df_train,df_validation,df_test], axis=0)
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_train
```

```python
df = df.drop(columns=['podcast', 'start', 'end'])
```

```python
features = df.columns[0:88]
targets = df.columns[88:]
features
```

```python
x = df.loc[:, features]
y = df.loc[:, targets]
```

```python
std_x = StandardScaler().fit_transform(x)
std_x.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=50)
principalComponents = pca.fit_transform(std_x)
principalDf = pd.DataFrame(data = principalComponents)
```
